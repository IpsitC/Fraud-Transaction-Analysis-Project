# Credit Card Fraud Transaction - Exploratory Data Analysis (EDA)

## Overview
This notebook explores the **Credit Card Fraud Transaction   dataset**.  
The goal is to analyze transaction patterns, compare fraud vs normal transactions, and identify key trends useful for fraud prevention.

```python
# Transaction Amount Analysis
# Let's compare the distribution of amounts for fraud vs normal.

import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv("../data/creditcard1.csv")
fraud = df[df['Class'] == 1]
normal = df[df['Class'] == 0]

# Fraud distribution pie chart
plt.pie([len(fraud), len(normal)], labels=["Fraud", "Normal"], autopct='%1.1f%%')
plt.title("Fraud vs Normal Transactions")
plt.show()
```

```python
# Fraud Rate by Transaction Amount
# We check which amount ranges have higher fraud probability.

import numpy as np
df['AmountBin'] = pd.cut(df['Amount'], bins=np.linspace(0, 2500, 20))  
fraud_rate = df.groupby('AmountBin')['Class'].mean()
fraud_rate.plot(kind='bar', figsize=(12,5))
plt.title("Fraud Rate by Transaction Amount")
plt.ylabel("Fraud Probability")
plt.show()
```

```python
# Summary Statistics
# Finally, let's compare fraud vs normal transactions in detail.

print("Normal Transactions Summary:")
print(normal['Amount'].describe())

print("\nFraud Transactions Summary:")
print(fraud['Amount'].describe())
```
