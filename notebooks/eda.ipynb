# Credit Card Fraud Transaction - Exploratory Data Analysis (EDA)

## Overview
This notebook explores the **Credit Card Fraud Transaction   dataset**.  
The goal is to analyze transaction patterns, compare fraud vs normal transactions, and identify key trends useful for fraud prevention.

```python
# Transaction Amount Analysis
# Let's compare the distribution of amounts for fraud vs normal.

import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv("../data/creditcard1.csv")
fraud = df[df['Class'] == 1]
normal = df[df['Class'] == 0]

# Fraud distribution pie chart
plt.pie([len(fraud), len(normal)], labels=["Fraud", "Normal"], autopct='%1.1f%%')
plt.title("Fraud vs Normal Transactions")
plt.show()
```

```python
# Transactions Over Time
# We check time trends to see when fraud is more likely.

plt.figure(figsize=(10,5))
plt.hist(df[df['Class']==0]['Time'], bins=50, alpha=0.5, label="Normal")
plt.hist(df[df['Class']==1]['Time'], bins=50, alpha=0.5, label="Fraud")
plt.legend()
plt.title("Fraud vs Normal Transactions Over Time")
plt.xlabel("Time (seconds)")
plt.ylabel("Count")
plt.show()
```
